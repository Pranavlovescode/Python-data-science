## Tutorial for matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Diabetes_prediction.csv')
plt.scatter(df['BMI'],df['Age'],c='purple')
plt.title('BMI vs Age')
plt.xlabel('BMI')
plt.ylabel('Age')
plt.show()
```

```python
# Histogram with default argument
plt.hist(df['Glucose'])
plt.show()
# Histogram with additional argument
plt.hist(df['Glucose'],color='red',edgecolor='white',bins=5)
plt.show()

```

```python
# Bar plot
# print(df.info())
df['BMI']=df['BMI'].astype('category')
# print(df.info())
plt.bar(df['Insulin'],df['Age'])
plt.xlabel('Insulin consumption')
plt.ylabel('Age')
plt.title('The bar graph of Insulin consumption to the age')
plt.show()

```
