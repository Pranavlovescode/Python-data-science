## Tutorial for matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('CAR DETAILS FROM CAR DEKHO.csv')
plt.scatter(df['selling_price'],df['fuel'],c='purple')
# plt.title('BMI vs Age')
# plt.xlabel('BMI')
# plt.ylabel('Age')
plt.show()
```

```python
# Histogram with default argument
plt.hist(df['owner'])
plt.show()
# Histogram with additional argument
plt.hist(df['fuel'],color='red',edgecolor='white',bins=5)
plt.show()

```

## Seaborn library

```python
import pandas as pd

# Assuming 'df' is your DataFrame
# df['fuel'] = pd.to_numeric(df['fuel'], errors='coerce')  # Convert 'fuel' column to numeric
# df['selling_price'] = pd.to_numeric(df['selling_price'], errors='coerce')  # Convert 'selling_price' column to numeric

# Now, you can proceed with plotting
import seaborn as sns
sns.set(style='darkgrid')
sns.regplot(x=df['fuel'], y=df['selling_price'])
```

```python
sns.lmplot(x='km_driven',y='selling_price',data=df,fit_reg=False,hue='year',legend=True,palette='Set1')
```

```python
import pandas as pd
import seaborn as sns

print(df)
sns.displot(df['km_driven'],kde=True)
```

```python
# Group bar plot
sns.countplot(x='km_driven',data=df,hue='transmission',palette='dark:green')
```

```python
# Box and Whiskers plot
sns.boxplot(y=df['selling_price'])
```

```python
sns.boxplot(x=df['transmission'],y=df['selling_price'])
```

```python
# Grouped box and whiskers plot
sns.boxplot(x='transmission',y='selling_price',hue='fuel',data=df)
```
